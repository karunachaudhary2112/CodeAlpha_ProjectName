```python
# This cell downloads the target webpage and shows the head of extracted table-like data.
# It uses requests and BeautifulSoup to parse the page and extract tables and links.
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://www.codealpha.tech"
resp = requests.get(url, timeout=15)
resp.encoding = resp.apparent_encoding
soup = BeautifulSoup(resp.text, "html.parser")

# Extract tables
tables = pd.read_html(resp.text) if len(soup.find_all('table'))>0 else []

# Extract main headings and links
headings = [h.get_text(strip=True) for h in soup.find_all(['h1','h2','h3'])]
links = [a.get('href') for a in soup.find_all('a', href=True)]

print(headings[:20])
print(len(links))

# If tables found, show head of first
if tables:
    df0 = tables[0]
    print(df0.head())
else:
    # Create a small dataframe of links and headings
    df_small = pd.DataFrame({'text': [a.get_text(strip=True) for a in soup.find_all('a', href=True)], 'href': links})
    print(df_small.head())

# Save page HTML for inspection
with open('codealpha_homepage.html', 'w', encoding='utf-8') as f:
    f.write(resp.text)

print('saved file: codealpha_homepage.html')
```
